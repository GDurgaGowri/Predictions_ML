```python
import tensorflow as tf
import tensorflow.keras.preprocessing.text as kpt
from tensorflow.keras.datasets import imdb
from nltk.corpus import stopwords
import nltk
import string
import pandas as pd

# Download stopwords
nltk.download('stopwords')

# Load IMDb dataset
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

# Load IMDb word index
word_index = imdb.get_word_index()

# Reverse mapping from index to word
index_to_word = {index + 3: word for word, index in word_index.items()}
index_to_word[0] = "[PAD]"
index_to_word[1] = "[START]"
index_to_word[2] = "[UNK]"
index_to_word[3] = "[UNUSED]"

# Function to decode reviews
def decode_review(encoded_review):
    return " ".join([index_to_word.get(i, "[UNK]") for i in encoded_review])

# Function to clean text
def clean_text(text):
    stop_words = set(stopwords.words("english"))
    text = text.lower()  # Convert to lowercase
    text = text.translate(str.maketrans("", "", string.punctuation))  # Remove punctuation
    words = text.split()
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return " ".join(words)

# Decode & clean reviews
train_texts = [clean_text(decode_review(x)) for x in train_data]
test_texts = [clean_text(decode_review(x)) for x in test_data]

# Convert to DataFrame
df_train = pd.DataFrame({"review": train_texts, "sentiment": train_labels})
df_test = pd.DataFrame({"review": test_texts, "sentiment": test_labels})

print("✅ IMDb Data Preprocessed Successfully!")
print(df_train.head())  # Show sample processed reviews
```

```python
df_combined = pd.concat([df_imdb, df_youtube], ignore_index=True)

# Shuffle dataset to avoid bias
df_combined = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the merged dataset
df_combined.to_csv("final_movie_reviews.csv", index=False)

print("✅ IMDb and YouTube datasets merged successfully!")
print(df_combined.head())
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
import numpy as np

# Define maximum vocabulary size and sequence length
VOCAB_SIZE = 10000  # Adjust based on dataset size
MAX_LENGTH = 200    # Adjust based on average review length

# Initialize Tokenizer
tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token="<OOV>")
tokenizer.fit_on_texts(df_combined["review"])

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(df_combined["review"])

# Pad sequences to make them uniform
padded_sequences = pad_sequences(sequences, maxlen=MAX_LENGTH, padding='post', truncating='post')

# Convert sentiment labels to NumPy array
labels = np.array(df_combined["sentiment"])

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)

print("✅ Text Preprocessing Completed Successfully!")
print(f"Training Samples: {len(X_train)} | Testing Samples: {len(X_test)}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense

# Define CNN model
cnn_model = Sequential([
    Embedding(VOCAB_SIZE, 128, input_length=MAX_LENGTH),
    Conv1D(128, 5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile CNN model
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train CNN model
cnn_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))

print("✅ CNN Model Training Completed!")
```

```python
from tensorflow.keras.layers import SimpleRNN

# Define RNN model
rnn_model = Sequential([
    Embedding(VOCAB_SIZE, 128, input_length=MAX_LENGTH),
    SimpleRNN(128, activation='relu', return_sequences=False),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile RNN model
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.layers import LSTM

# Define LSTM model
lstm_model = Sequential([
    Embedding(VOCAB_SIZE, 128, input_length=MAX_LENGTH),
    LSTM(128, return_sequences=False),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile LSTM model
lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.layers import Bidirectional

# Define BiLSTM model
bilstm_model = Sequential([
    Embedding(VOCAB_SIZE, 128, input_length=MAX_LENGTH),
    Bidirectional(LSTM(128, return_sequences=False)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile BiLSTM model
bilstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history_cnn = cnn_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
history_rnn = rnn_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
history_lstm = lstm_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
history_bilstm = bilstm_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt

# Function to plot accuracy and loss
def plot_history(history, model_name):
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # Accuracy Plot
    axs[0].plot(history.history['accuracy'], label='Train Accuracy', color='blue')
    axs[0].plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
    axs[0].set_title(f'{model_name} Accuracy')
    axs[0].set_xlabel('Epochs')
    axs[0].set_ylabel('Accuracy')
    axs[0].legend()

    # Loss Plot
    axs[1].plot(history.history['loss'], label='Train Loss', color='blue')
    axs[1].plot(history.history['val_loss'], label='Validation Loss', color='red')
    axs[1].set_title(f'{model_name} Loss')
    axs[1].set_xlabel('Epochs')
    axs[1].set_ylabel('Loss')
    axs[1].legend()

    plt.show()

# Plot for each model
plot_history(history_cnn, "CNN")
plot_history(history_rnn, "RNN")
plot_history(history_lstm, "LSTM")
plot_history(history_bilstm, "BiLSTM")
```

```python
import time

# Function to train and time models
def train_and_time(model, X_train, y_train, X_test, y_test, epochs=5, batch_size=32):
    start_time = time.time()
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test))
    end_time = time.time()

    training_time = end_time - start_time
    final_accuracy = history.history['val_accuracy'][-1]

    return final_accuracy, training_time

# Train models and record time
cnn_acc, cnn_time = train_and_time(cnn_model, X_train, y_train, X_test, y_test)
rnn_acc, rnn_time = train_and_time(rnn_model, X_train, y_train, X_test, y_test)
lstm_acc, lstm_time = train_and_time(lstm_model, X_train, y_train, X_test, y_test)
bilstm_acc, bilstm_time = train_and_time(bilstm_model, X_train, y_train, X_test, y_test)

# Store results
results = {
    "Model": ["CNN", "RNN", "LSTM", "BiLSTM"],
    "Accuracy": [cnn_acc, rnn_acc, lstm_acc, bilstm_acc],
    "Training Time (seconds)": [cnn_time, rnn_time, lstm_time, bilstm_time]
}

import pandas as pd
df_results = pd.DataFrame(results)
print(df_results)
```
